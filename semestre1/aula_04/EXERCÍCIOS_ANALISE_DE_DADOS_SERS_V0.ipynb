# **Exemplo**

Analisar dados de consumidores de energia elétrica usando Python.


```python
# Importar biblioteca de análise de dados ----> Pandas
import pandas as pd
```

```python
# Carregar os dados criando uma tabela
tabela = pd.read_csv('/content/household_power_consumption.csv')
```

```python
# Visualizar primeiros registros da tabela
tabela.head()
```

```python
# Levantando estatísticas descritivas
tabela.describe()
```

```python
# Colunas
tabela.columns
```

```python
# Calculando a média aritmética de
# uma coluna Voltage (Tensão ----> em Volts)
# mean() ---> calcula média
# median() ---> calcula mediana
# std() ---> calcula desvio padrão
# round() ---> arredonda
media_tensao = tabela['Voltage'].mean()
mediana_tensao = tabela['Voltage'].median()
desvio_padrao_tensao = tabela['Voltage'].std()
# media_tensao, mediana_tensao, desvio_padrao_tensao
print(f'A média aritmética da coluna Voltage é {media_tensao:.3f}V')
print(f'A mediana da coluna Voltage é {mediana_tensao:.3f}V')
print(f'O desvio padrão da coluna Voltage é {desvio_padrao_tensao:.3f}V')
```

```python
# Amostra de 10% dos dados originais
tabela_amostra = tabela.sample(frac=0.1)
```

```python
tabela.shape, tabela_amostra.shape
```

```python
# Dados da amostra
media_tensao_amostra = tabela_amostra['Voltage'].mean()
mediana_tensao_amostra = tabela_amostra['Voltage'].median()
desvio_padrao_tensao_amostra = tabela_amostra['Voltage'].std()
# media_tensao_amostra, mediana_tensao_amostra, desvio_padrao_tensao_amostra
print(f'A média aritmética da coluna Voltage é {media_tensao_amostra:.3f}V')
print(f'A mediana da coluna Voltage é {mediana_tensao_amostra:.3f}V')
print(f'O desvio padrão da coluna Voltage é {desvio_padrao_tensao_amostra:.3f}V')
```
