## **Fórmulas Essenciais**

Para resolver os exercícios, utilize as seguintes fórmulas:

**Potência Ativa (P)** - Medida em Watts (W) ou Quilowatts (kW):
$$P = \frac{P_{util}}{N}$$
Onde $P_{util}$ é a potência mecânica útil e $N$ é o rendimento do motor.

**Potência Aparente (S)** - Medida em Volt-Amperes (VA) ou Quilovolt-Amperes (kVA):
$$S = \frac{P}{FP}$$
Onde $FP$ é o fator de potência.

**Potência Reativa (Q)** - Medida em Volt-Amperes Reativos (VAr) ou Quilovolt-Amperes Reativos (kVAr):
$$Q = \sqrt{S^2 - P^2}$$

## **Questão 1: Infraestrutura de Resfriamento (2,0 pontos)**

Você trabalha em uma consultoria de eficiência energética e foi contratado para analisar o consumo de energia de um pequeno data center. O cliente precisa entender quanto de potência elétrica seu sistema de ventilação consome da rede.

O rendimento é a razão entre a potência útil que o motor entrega e a potência elétrica que consome. Por exemplo, um motor com 0,79 de rendimento aproveita 79% da energia elétrica para trabalho útil, perdendo 21% em calor e atrito.

Primeiro, você fará o levantamento atual do sistema. Depois, simulará dois cenários: um com equipamento mais moderno e outro com melhorias na qualidade da energia.

Consulte DADOS_MOTORES.md - Questão 1 para obter os valores.

```python
# 1.1 LEVANTAMENTO ATUAL - Entrada de dados
# Insira os valores do arquivo DADOS_MOTORES.md - Questão 1

P_util_1 = 1.5
N_1 = 0.78
FP_1 = 0.82
```

```python
# 1.2 Calcule a Potência Ativa (P) do sistema atual
# Nome da variável obrigatório: P1
# Fórmula: P = P_util / N

P1 = P_util_1 / N_1

print(f"A potência ativa é: {P1:.2f}")
```

```python
# 1.3 Calcule a Potência Aparente (S) do sistema atual
# Nome da variável obrigatório: S1
# Fórmula: S = P / FP

S1 = P1 / FP_1

print(f"A potência Aparente do motor é: {S1:.2f}")
```

```python
# 1.4 Calcule a Potência Reativa (Q) do sistema atual
# Nome da variável obrigatório: Q1
# Fórmula: Q = sqrt(S² - P²)
# Importante: Use (S² - P²)**0.5 para calcular a raiz quadrada

Q1 = (S1**2 - P1**2)**0.5
print(f"A Potência Reativa do motor é: {Q1:.2f}")
```

```python
# 1.5 RESULTADO DO LEVANTAMENTO ATUAL
# Crie um print com f-string exibindo os valores de P, S e Q
# Exemplo: print(f'Potência Ativa: {P1:.4f} kW')

print(f"""A Potência Ativa é: {P1:.4}kW
A Potência Aparente é: {S1:.4f}kW
A Potência Reativa é: {Q1:.4f}kW""")
```

## **Questão 4: Análise Comparativa de Motores (3,0 pontos)**

Uma empresa de logística precisa escolher entre dois fornecedores de motores para suas operações. Ambos oferecem motores com a mesma potência útil, mas com características diferentes. Você será responsável por comparar o consumo de energia de ambas as opções e recomendar a melhor escolha.

Além da análise inicial, você simulará melhorias em ambos os motores para entender o impacto de investimentos em rendimento energético.

Consulte DADOS_MOTORES.md - Questão 4 para obter os valores.

```python
# 4.1 MOTOR A - Entrada de dados
# Insira os valores do arquivo DADOS_MOTORES.md - Questão 4 - Motor A

P_util_A = 5.5
N_A = 0.85
FP_A = 0.86
```

```python
# 4.2 MOTOR B - Entrada de dados
# Insira os valores do arquivo DADOS_MOTORES.md - Questão 4 - Motor B

P_util_B = 5.5
N_B = 0.83
FP_B = 0.85
```

```python
# 4.9 Calcule a Potência Reativa do Motor B
# Nome da variável obrigatório: QB
# Fórmula: Q = sqrt(S² - P²)

QB = (SB**2 - PB**2)**0.5
```
