# **Exemplo - Aula 01 - Demanda e Potência**

```python
# Entrada de dados
PU = int(input('Insira a potência útil do motor: '))
n = float(input('Insira o valor do rendimento (de 0 a 1): '))
```

```python
# Cálculo da potência ativa
# n =  PU / P
P = PU / n
```

```python
print(f'A potência ativa do motor é: {int(P)} W')
```
